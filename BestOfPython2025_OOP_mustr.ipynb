### âºï¸ Abstrakce a interface

DÄ›diÄnost je fajn ale neÅ™eÅ¡Ã­ vÅ¡echny naÅ¡e problÃ©my. ObÄas potÅ™ebujeme mÄ›nit schopnosti a morfologii danÃ©ho objektu, podle toho co mÃ¡ umÄ›t a jak mÃ¡ pÅ¯sobit.
````python
from abc import ABC, abstractmethod

# AbstraktnÃ­ tÅ™Ã­da pro vÅ¡echny zbranÄ›
class Weapon(ABC):

    # AbstraktnÃ­ metoda pro Ãºder (napÅ™. pÄ›stÃ­, palicÃ­)
    @abstractmethod
    def bash(self):
        pass

    # AbstraktnÃ­ metoda pro Å¡vihnutÃ­ (napÅ™. meÄem)
    @abstractmethod
    def swing(self):
        pass

# TÅ™Ã­da reprezentujÃ­cÃ­ meÄ, dÄ›dÃ­ z Weapon
class Sword(Weapon):

    def bash(self):
        # MeÄ nenÃ­ ideÃ¡lnÃ­ na Ãºder, ale lze jÃ­m zasÃ¡hnout rukojetÃ­
        print("You bash with the pommel of your sword.")

    def swing(self):
        # TypickÃ½ Ãºtok pro meÄ
        print("You swing your sword in a clean arc.")

# TÅ™Ã­da reprezentujÃ­cÃ­ palici, dÄ›dÃ­ z Weapon
class Club(Weapon):

    def bash(self):
        # HlavnÃ­ zpÅ¯sob Ãºtoku palicÃ­
        print("You bash with your heavy club.")

    def swing(self):
        # I palici lze rozmÃ¡chnout
        print("You swing your club with brute force.")

````
